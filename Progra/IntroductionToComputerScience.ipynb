```python
####################
## EXAMPLE: perfect squares
####################
ans = 0
neg_flag = False
x = int(input("Enter an integer: "))
if x < 0:
    neg_flag = True
    x = -x
while ans**2 < x:
    ans = ans + 1
if ans**2 == x:
    if neg_flag:
        print("Square root of", -x, "is", ans,"i")
    elif neg_flag == False:
        print("Square root of", x, "is", ans)
else:
    if neg_flag:
        print(-x, "is not a perfect square")
    elif neg_flag == False:
        print(x, "is not a perfect square")
```
